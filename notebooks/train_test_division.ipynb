```python
import random
from collections import Counter, defaultdict
from sklearn.model_selection import train_test_split

# Load CoNLL data into list of (tokens, labels)
def load_conll_data(path):
    sentences = []
    sentence = []
    with open(path, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            if not line:
                if sentence:
                    sentences.append(sentence)
                    sentence = []
                continue
            parts = line.split()
            if len(parts) >= 2:
                token, label = parts[0], parts[-1]
                sentence.append((token, label))
    if sentence:
        sentences.append(sentence)
    return sentences

# Get a representative label per sentence (majority class or any present B- label)
def get_sentence_label_distribution(sentences):
    labels_per_sentence = []
    for s in sentences:
        labels = [label for _, label in s if label != 'O']
        if not labels:
            labels = ['O']
        # We'll take the first non-O label as representative (or majority label)
        labels_per_sentence.append(labels[0])
    return labels_per_sentence

# Save in CoNLL format
def save_conll(sentences, path):
    with open(path, 'w', encoding='utf-8') as f:
        for s in sentences:
            for token, label in s:
                f.write(f"{token} {label}\n")
            f.write("\n")

# --- Main script ---
conll_path = "synthetic_v1.conll"
sentences = load_conll_data(conll_path)
labels = get_sentence_label_distribution(sentences)

# Stratified train/test split
train_sentences, test_sentences = train_test_split(
    sentences, test_size=0.2, stratify=labels, random_state=42
)

# Save
save_conll(train_sentences, "synthetic_train_v1.conll")
save_conll(test_sentences, "synthetic_test_v1.conll")

print(f"Train samples: {len(train_sentences)}")
print(f"Test samples: {len(test_sentences)}")
```

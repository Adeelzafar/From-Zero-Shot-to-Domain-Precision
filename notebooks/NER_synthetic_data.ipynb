```python
# Template generator
base_templates = [
    "The {EN} in the {LOC} {FT} due to {DP} (code {FC}). {VH} showed {MV}. Symptom: {CS}. Action: {MM}.",
    "{VH} reported {FT} in {EN} at the {LOC}. Detected: {DP} with code {FC}. Resolved: {MM}.",
    "Complaint: {CS}. Diagnosis: {FT} on {EN} located {LOC}. Fault Code: {FC}. Sensor: {MV}. Fixed by: {MM}.",
    "Observed {DP} and {FT} on the {EN} ({FC}) in {VH}. Location: {LOC}. Performed: {MM}.",
    "After {TD}, user noted {CS}. Cause was {DP} on {EN}, code {FC}, at {LOC}. Maintenance: {MM}.",
    "{VH} reported an issue: {EN} {FT} causing {DP}. {MM}. Fault code(s): {FC}. Location: {LOC}. Measurement(s): {MV}. Duration: {TD}. Symptom(s): {CS}.",
    "During {TD}, {VH} experienced {CS}. {EN} showed {DP} with {FT}. Actions: {MM}. Fault codes: {FC}. Location: {LOC}. Measurements: {MV}.",
    "Vehicle {VH} has problems with {EN} ({DP}): {FT}. Customer reported {CS}. Maintenance performed: {MM}. Fault code(s): {FC}. Location: {LOC}. Measurements: {MV}."
]
```

```python
import random
import pandas as pd
from tqdm import tqdm

# -----------------------------
# Generate dynamic templates
# -----------------------------
def generate_templates(n=200):
    templates = []
    for _ in range(n):
        t = random.choice(base_templates)
        if random.random() > 0.5:
            t = t.replace("Fault code", "code")
        if random.random() > 0.5:
            t += " Observed over duration: {TD}."
        templates.append(t)
    return templates

templates = generate_templates(200)

# -----------------------------
# Generate single NER record
# -----------------------------
def generate_ner_record(template):
    # Pick multiple items for realism
    EN_list = random.sample(equipment, k=random.randint(1, 2))
    DP_list = random.sample(properties, k=random.randint(1, 2))
    FT_list = random.sample(faults, k=random.randint(1, 3))
    FC_list = random.sample(fault_codes, k=random.randint(1, 2))
    MM_list = random.sample(maintenance, k=random.randint(1, 2))
    LOC_list = random.sample(locations, k=random.randint(1, 2))
    MV_list = random.sample(measurements, k=random.randint(1, 2))
    VH = random.choice(vehicles)
    TD = random.choice(time_durations)
    CS_list = random.sample(customer_symptoms, k=random.randint(1, 2))

    # Join lists into text
    EN = ", ".join(EN_list)
    DP = ", ".join(DP_list)
    FT = ", ".join(FT_list)
    FC = ", ".join(FC_list)
    MM = "; ".join(MM_list)
    LOC = ", ".join(LOC_list)
    MV = ", ".join(MV_list)
    CS = ", ".join(CS_list)

    text = template.format(
        EN=EN, DP=DP, FT=FT, FC=FC, MM=MM,
        LOC=LOC, MV=MV, VH=VH, TD=TD, CS=CS
    )

    # Generate NER entities for all items
    entities = []
    for label, values in [
        ("EquipmentName", EN_list),
        ("DeviceProperties", DP_list),
        ("FaultType", FT_list),
        ("FaultCode", FC_list),
        ("MaintenanceMethod", MM_list),
        ("VehicleComponentLocation", LOC_list),
        ("MeasurementValue", MV_list),
        ("VehicleMakeModel", [VH]),
        ("TimeDuration", [TD]),
        ("CustomerReportedSymptom", CS_list)
    ]:
        for value in values:
            start = text.find(value)
            if start >= 0:
                entities.append({
                    "start": start,
                    "end": start + len(value),
                    "label": label
                })

    return {"text": text, "entities": entities}

```

```python

# -----------------------------
# Generate full dataset
# -----------------------------
def generate_dataset(n_records=10000):
    records = []
    for _ in tqdm(range(n_records)):
        template = random.choice(templates)
        record = generate_ner_record(template)
        records.append(record)
    return records

```

```python
# -----------------------------
# Generate and save dataset
# -----------------------------
dataset = generate_dataset(10000)
df = pd.DataFrame(dataset)
df.to_json("synthetic_vehicle_ner_data_realistic.json", orient="records", indent=2)
df.to_csv("synthetic_vehicle_ner_data_realistic.csv", index=False)

print("✅ Realistic synthetic NER dataset created with 10,000 examples.")
```

```python
import json

# Load your original synthetic dataset
with open("adversarial_synthetic_data.json", "r") as f:
    raw_data = json.load(f)

label_studio_data = []

for example in raw_data:
    text = example["text"]
    entities = example.get("entities", [])

    annotation = {
        "data": {
            "text": text
        },
        "annotations": [
            {
                "result": []
            }
        ]
    }

    for entity in entities:
        start = entity["start"]
        end = entity["end"]
        label = entity["label"]
        entity_text = text[start:end]

        annotation["annotations"][0]["result"].append({
            "from_name": "label",
            "to_name": "text",
            "type": "labels",
            "value": {
                "start": start,
                "end": end,
                "text": entity_text,
                "labels": [label]
            }
        })

    label_studio_data.append(annotation)

# Save to Label Studio compatible JSON
with open("label_studio_advarsarial_dataset.json", "w") as out_file:
    json.dump(label_studio_data, out_file, indent=2)
```
